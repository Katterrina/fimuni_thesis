```python
pytepfit_AUC = pd.read_csv(paths.path('interim/pyTepFit_empirical/not_over_threshold_nan/response_AUC_and_communication_metrics_M1_pyTepFit_200ms_thr7.csv')).set_index('label')
pytepfit_AUC.index = pytepfit_AUC.index.map(lambda x: x[:-3])
pytepfit_AUC
```

```python
mapping_schaefer_schaefer = pd.read_csv(paths.path(f'external/schaefer_parcellation_centroids/ROI_MAPPING_pytepfit.csv'),index_col=0)
pytepfit_AUC = pd.merge(pytepfit_AUC,mapping_schaefer_schaefer,left_on="label",right_on='roi_name')
pytepfit_AUC[["response","ED","roi_name","roi_name_csv"]]
```

```python
mapping = pd.read_csv(paths.path('interim/DICE_Glasser_Schaefer200_labelMapping_greedy.csv'),index_col=0)
mapping = mapping[["regionName","Yeo17","Schaefer200","regionLongName"]]
mapping
```

```python
def reorder_label(x):
    if x[0] == 'L':
        return x[2:]+"_L"
    else:
        return x[2:]+"_R"

ftract_probabilities.index = ftract_probabilities.index.map(reorder_label)
```

```python
set(ftract_probabilities.index) == set(mapping["regionName"])
```

```python
merged_ftract_dice = pd.merge(mapping,ftract_probabilities,left_on="regionName",right_on="label")
merged_all = pd.merge(merged_ftract_dice,pytepfit_AUC,left_on="Schaefer200",right_on="roi_name_csv",suffixes=["","_TMS"])
merged_all[["regionName","Yeo17","Schaefer200","response","ED","response_TMS","ED_TMS"]]
```

```python
merged_all["color17"] = merged_all["Yeo17"].apply(lambda x: int(x.split("_")[0]))
color7_dict= {'Default':'yellow', 'Limbic':'blue','SalVentAttn':'red',  'DorsAttn':'green','Vis':'purple','Cont':'orange', 'SomMot':'pink'}
merged_all["color7"] = merged_all["Schaefer200"].apply(lambda x: color7_dict[x.split("_")[2]])
merged_all[["regionName","Yeo17","Schaefer200","response","ED","response_TMS","ED_TMS","color17","color7"]]
```

```python
legend_patches = []
for c in color7_dict:
    legend_patches.append(mpatches.Patch(color=color7_dict[c], label=c))

def minmax_norm(x):
    return (x - np.nanmin(x)) / (np.nanmax(x)-np.nanmin(x))

def plot_two_columns(df,col1,col2,label,normalize=True,log_axes=False,labels=False,corr_line=None):
    if normalize:
        a1 = minmax_norm(np.array(df[col1]))
        a2 = minmax_norm(np.array(df[col2]))
    else:
        a1 = np.array(df[col1])
        a2 = np.array(df[col2])
        
    plt.style.use('ggplot')
    fig,ax = plt.subplots(figsize=(7,7))

    if corr_line:
        ax.plot(corr_line[0],corr_line[1],c="gray")
    
    ax.scatter(a1,a2,c=df["color7"])
    ax.legend(handles=legend_patches,loc='lower right',ncols=2)

    ax.set_xlabel(col1+"_F-Tract")
    ax.set_ylabel(col2)


    if log_axes:
        ax.set_yscale('log')
        ax.set_xscale('log')
    
    if labels and not log_axes:
        texts = []

        for i, coor in enumerate(zip(a1,a2)):
            x,y = coor
            if np.isnan(x) or np.isnan(y):
                continue

            texts.append(ax.text(x,y,df[label][i],fontsize='x-small'))

        adjust_text(texts, force_text=(0.5,0.5),expand=(1.2,1.2),expand_axes=True,time_lim=5,arrowprops=dict(arrowstyle='-', color='black', lw=0.5))


plot_two_columns(merged_all,"ED","ED_TMS","regionName",normalize=False,corr_line = [[0,140],[0,140]])
```
