Data processing: Shared data have been obtained after applying the following
pre-processing steps to raw TMS-EEG recordings:
1) visual rejection of artifact-contaminated trials and channels;
2) high-pass filtering at 1 Hz;
3) epoching between -400 and +400 ms around TMS pulses;
4) re-referencing to the average reference and baseline correction (from -400 to -5
ms);
5) Independent Component Analysis to reduce ocular and muscular artifacts;
6) low-pass filtering at 45 Hz;
7) downsampling to 725 Hz;
8) interpolation of artifact-contaminated channels using spherical interpolation.
